```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean()
```

```python
avg_scores = df.groupby("test preparation course")[
    ["math score", "reading score", "writing score"]
].mean()

avg_scores.plot(kind="bar")
plt.title("Average Scores by Test Preparation Course")
plt.ylabel("Average Score")
plt.show()
```

```python
df.groupby("lunch")["math score"].mean().plot(kind="bar")
plt.title("Average Math Score by Lunch Type")
plt.ylabel("Average Math Score")
plt.show()
```

### Insights

Insight 1: Students who completed the test preparation course have higher average scores across math, reading, and writing compared to those who did not.

Insight 2: Reading and writing scores show a larger improvement with test preparation than math scores.

Insight 3: Students with standard lunch tend to have higher average math scores than students with free/reduced lunch.
